```python
# Load the Excel file
workbook = openpyxl.load_workbook('/content/drive/MyDrive/CS210 Code Part/Visualization/Milliyetlerine-gore-gelen-ziyaretciler-2022.xlsx')

# Select the desired sheet
sheet = workbook.worksheets[0]

# Find the "Toplam" row and retrieve the data
for row in sheet.iter_rows(values_only=True):
    if row[0] == 'Toplam':
        toplam_data = row[1:]
        break


# Find the row containing the years and retrieve the data
for row in sheet.iter_rows(values_only=True):
    if row[0] == 'MİLLİYET':
        years = row[1:]
        break


# Convert the 'toplam_data' values from string to integer (if necessary)
total_data = []
for data in toplam_data:
    if isinstance(data, str):
        data = int(data.replace(' ', '').replace(',', ''))
    total_data.append(data)


# Create the figure and axes
fig, ax = plt.subplots()

# Plot the time series graph
ax.plot(years, total_data, marker='o', linestyle='-', color='blue', linewidth=2)

# Set the labels for the x-axis and y-axis
ax.set_xlabel('Years')
ax.set_ylabel('Tourist Number')

# Set the title of the graph
ax.set_title('Number of Tourists over Years')

# Rotate the x-axis labels for better readability
plt.xticks(rotation=45)

# Add gridlines
ax.grid(True, linestyle='--', alpha=0.5)

# Add a background color to the plot
ax.set_facecolor('#f7f7f7')

# Customize the tick labels font size
ax.tick_params(axis='both', which='major', labelsize=10)

# Remove the top and right spines
ax.spines['top'].set_visible(False)
ax.spines['right'].set_visible(False)

# Add data labels to each point
for x, y in zip(years, total_data):
    ax.text(x, y, f'{y/1e6:.0f}m', ha='center', va='bottom', fontsize=8)

# Format y-axis tick labels to display numbers in millions
ax.yaxis.set_major_formatter(ticker.FuncFormatter(lambda x, pos: f'{x/1e6:.0f}m'))

# Adjust the layout to prevent labels from being cut off
plt.tight_layout()

# Display the graph
plt.show()
```

```python
df = pd.read_excel('TUFE-bazlı-reel-efektif-döviz-kuru.xlsx')

# Extract the required column
column_name = 'TÜFE  Bazlı Reel Efektif Döviz Kuru (2003=100) '
column_data = df[column_name]


# Convert date strings to datetime format
df['Dönem'] = pd.to_datetime(df['Dönem'], format='%m/%y')

# Create the graph
fig, ax = plt.subplots(figsize=(10, 6))
plt.plot(df['Dönem'], column_data, marker='o', markersize=4, linestyle='-', linewidth=1)

# Customize the appearance
plt.xlabel('Time', fontsize=12)
plt.ylabel(column_name, fontsize=12)
plt.title('Shift in TUFE Bazlı Reel Efektif Döviz Kuru', fontsize=14)
plt.xticks(rotation=45, fontsize=10)
plt.yticks(fontsize=10)
plt.grid(True, linestyle='--', alpha=0.5)

# Set x-axis limits to include all years
plt.xlim(pd.Timestamp('1994-01-01'), pd.Timestamp('2023-12-31'))

# Add more detailed date labels for each year
years = pd.date_range(start='1994', end='2023', freq='YS')
ax.set_xticks(years)
ax.set_xticklabels(years.year, rotation=45, ha='right', fontsize=8)

plt.tight_layout()
plt.show()
```

```python
# Read the Excel file
df = pd.read_excel('Milliyetlerine-gore-gelen-ziyaretciler-2022.xlsx', header=2).dropna().drop_duplicates()


# Extract the required columns
nationality_column_index = 0
year_columns = df.columns[1:]  # Assuming the year columns start from the second column

# Calculate the total number of tourists for each nationality for each year
df['Total'] = df[year_columns].sum(axis=1)

# Sort the DataFrame by the total number of tourists in descending order
df_sorted = df.sort_values(by='Total', ascending=False)

# Select the top 30 nationalities
df_top30 = df_sorted.head(30)

# Exclude the 'Toplam' row if it exists
df_top30 = df_top30[df_top30.iloc[:, nationality_column_index] != 'Toplam']

# Filter out nationalities with less than 2% of the total
threshold = 0.02  # 2%
other_total = df_top30[df_top30['Total'] / df_top30['Total'].sum() < threshold]['Total'].sum()
other_nationalities = df_top30[df_top30['Total'] / df_top30['Total'].sum() < threshold].iloc[:, nationality_column_index]
df_top30 = df_top30[df_top30['Total'] / df_top30['Total'].sum() >= threshold]

# Add the 'Other' category to the DataFrame
df_other = pd.DataFrame({df.columns[nationality_column_index]: ['Other'], 'Total': [other_total]})
df_top30 = df_top30.append(df_other, ignore_index=True)

# Prepare the data for the pie chart
nationalities = df_top30[df.columns[nationality_column_index]]
totals = df_top30['Total']

# Create the pie chart
fig, ax = plt.subplots(figsize=(8, 8))
explode = [0.1] + [0] * (len(nationalities) - 1)  # Explode the first slice for emphasis
wedges, labels, autopct = plt.pie(
    totals,
    labels=nationalities,
    explode=explode,
    startangle=90,
    autopct='%1.1f%%',
    pctdistance=0.85,  # Adjust the percentage label distance from the center
    wedgeprops={'linewidth': 1.0, 'edgecolor': 'white'}  # Add a white edge to the wedges
)

# Adjust label properties
plt.setp(labels, fontsize=10)  # Set font size for the labels

# Customize the appearance
plt.title('Top Nationalities of Tourists')
plt.axis('equal')  # Equal aspect ratio ensures that pie is drawn as a circle

plt.show()

```

```python
import pandas as pd
import matplotlib.pyplot as plt

# Read the CSV file and skip unnecessary rows
df = pd.read_csv('gelis-nedenine-gore-cikis-yapan-z.csv').dropna()

# Exclude 'Toplam' column
df = df.iloc[:, :-1]

# Extract the years and spending types
years = df.columns[0]
spending_types = df.columns[1:]

# Convert data to numeric and divide by 1 million
numeric_columns = df.columns[1:]
df[numeric_columns] = df[numeric_columns].apply(lambda x: x.str.replace(' ', '').str.replace(',', '').astype(int) / 1000000)

# Drop rows with missing values
df.dropna(inplace=True)

# Filter data for years 2003-2011 and 2011-2019
df_2003_2011 = df[(df[years] >= 2003) & (df[years] <= 2011)]
df_2011_2019 = df[(df[years] >= 2011) & (df[years] <= 2019)]

# Prepare the data for plotting
data_2003_2011 = df_2003_2011.set_index(years)
data_2011_2019 = df_2011_2019.set_index(years)

# Calculate the maximum spending amount for both dataframes
max_spending = max(data_2003_2011.max().max(), data_2011_2019.max().max())

# Increase the figure size for both graphs
fig1, ax1 = plt.subplots(figsize=(12, 8))
fig2, ax2 = plt.subplots(figsize=(12, 8))

# Create the first graph for years 2003-2011
data_2003_2011.plot(kind='bar', ax=ax1)

# Set the plot labels and title for the first graph
ax1.set_xlabel('Years')
ax1.set_ylabel('Visiting Reasons in Millions')
ax1.set_title('Distribution of Visiting Reasons by Years (2003-2011)')

# Set the y-axis limit for the first graph
ax1.set_ylim(0, max_spending)

# Show the legend for the first graph
ax1.legend(title='Visiting Reasons', bbox_to_anchor=(1, 1))

# Create the second graph for years 2011-2019
data_2011_2019.plot(kind='bar', ax=ax2)

# Set the plot labels and title for the second graph
ax2.set_xlabel('Years')
ax2.set_ylabel('Visiting Reasons in Millions')
ax2.set_title('Distribution of Visiting Reasons by Years (2011-2019)')

# Set the y-axis limit for the second graph
ax2.set_ylim(0, max_spending)

# Show the legend for the second graph
ax2.legend(title='Visiting Reasons', bbox_to_anchor=(1, 1))

# Adjust the spacing
fig1.tight_layout()
fig2.tight_layout()

# Display the graphs
plt.show()
```
