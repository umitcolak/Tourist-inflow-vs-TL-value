    Hypothesis Testing
    

We got the data from the parts above.
After that we extract the rows where the data is above average to determine if there is a correlation between TL and turism in Turkey.
After that we apply randomized test where we pick a random year from 3 of the arrays that contain above average data. Then we check if in that year we have above average visitors, spendings and TL below the endex taken from "Tüfe bazlı reel efektif döviz kuru". If all 3 checks out true then we put a 1 in the test_array indicating that there is a correlation, otherwise we put a 0 indicating not. After that we calculate the p-value and compare it with the significance level (0.05). 

```python
import  random
import numpy as np


""" 
---> NULL: The depreciation of the Turkish lira leads to an increase in tourism in Turkey.
---> Altenative: The increase in the tourism in Turkey is not correlated with the depreciation of the Turkish lira.
"""

# get the rows that have incoming visitors above average
inc = general_table["Incoming Visitors"]
general_table_above_avg = pd.DataFrame(general_table[inc > inc.mean()])
years_incoming_visitor_above_avg = general_table_above_avg["Years"]

# get the rows that have average spending above average
spending = general_table["Average Spending"]
general_table_spending_above_avg = pd.DataFrame(general_table[spending > spending.mean()])

# get the rows with year when there are incoming visitors above the average
turist_years_above_avg = turist_years[turist_years['Years'].isin(years_incoming_visitor_above_avg)]
russia_ruble_above_avg = russia_ruble[russia_ruble['Years'].isin(years_incoming_visitor_above_avg)]
german_euro_above_avg = german_euro[german_euro['Years'].isin(years_incoming_visitor_above_avg)]
british_euro_above_avg = british_euro[british_euro['Years'].isin(years_incoming_visitor_above_avg)]
nationality_turist_years_above_avg = nationality_turist_years[nationality_turist_years.columns.intersection(years_incoming_visitor_above_avg)]
nationality_turist_years_above_avg.insert(0, 'MİLLİYET', nationality_turist_years["MİLLİYET"])



# getting the rows where the TL endex is below average
rate = TL["TÜFE  Bazlı Reel Efektif Döviz Kuru (2003=100) "]
TL_below = pd.DataFrame(TL[rate < rate.mean()])
TL_below_index = np.array(TL_below.index)
TL_below_index = TL_below_index.astype(np.int64)

# turning them into np arrays
general_spending_abv_avg = np.array(general_table_spending_above_avg["Years"])
general_abv_avg = np.array(general_table_above_avg["Years"])


 # Randomized test
test_array = []
# choose a random array from [above average spending, above average incoming visitors, TL below average endex]
# then choose a random year from that array
# Check if that year is in all 3 of the arrays listed above
# if yes append 1 to the test array which means there is a correlation between TL and turism
# else append 0 which means there is no correlation
for i in range(100):
    poss = random.randint(0,2)
    years = [general_spending_abv_avg, general_abv_avg, TL_below_index]
    year = years[poss][random.randint(0,len(years[poss])-1)]
    if year in general_abv_avg and year in general_spending_abv_avg and year in TL_below_index:
        test_array.append(1)
    else:
        test_array.append(0)
test_array = np.array(test_array)

# 1's are in the null hypothesis's favor
# 0's mean its wrong

p_value = len(test_array[np.where(test_array == 1)]) / len(test_array)
print("p-value: ", p_value)


alpha = 0.05

if p_value <= alpha:
  print("We reject Null Hypothesis")
else:
  print("We fail to reject Null Hypothesis!")
```
